```python
import pandas as pd
import pytest
from cat_regression import get_features_cat_regression
```

```python
def test_get_features_cat_regression_ok():
    df = pd.DataFrame({
        "color": ["rojo", "rojo", "azul", "azul"],
        "ventas": [100, 110, 10, 15]})
    resultado = get_features_cat_regression(
        df=df,
        target_col="ventas",
        pvalue=0.05)
    assert isinstance(resultado, list)
    assert "color" in resultado
```
